```python
# Primeiro, criar pastas de 2010 até 2021, separada por meses e dias
#Exemplo => 2021 -> 01 -> 20
#Criar estrutra no HD D:\
```

```python
import os
#from pathlib import Path
from datetime import datetime      
import shutil
```

```python
path='D:/'
path_mural='D:/MURAL/'
```

```python
#rotina para criar as pastas do ano, meses e dia
def criarPastas(anos = range(2010, 2022, 1),path=path_mural):
    
    for ano in anos:
        for mes in range(1,13,1):
            for dia in range(1,32,1):
                pasta_ano_mes_dia = os.path.join(path,str(ano),str(mes),str(dia))
                os.makedirs(pasta_ano_mes_dia)
                
```

```python
criarPastas()
```

```python
def selecionarPastaAnalisada(raiz,caminho_fotos):
    pastaAnalisaFotos=os.path.join(path,caminho_fotos)
    os.chdir(pastaAnalisaFotos)
    return pastaAnalisaFotos
```

```python
caminho=selecionarPastaAnalisada(path,'Galeria-Fotos/Minhas/')
```

```python
def movimentarFotosPastas(caminho=caminho,path_mural=path_mural):
    lista_arquivos = os.listdir(caminho)
    erro=[]
    for arquivo in lista_arquivos:
        ano_mes_dia= datetime.fromtimestamp(os.path.getctime(arquivo))
        ano = ano_mes_dia.year
        mes = ano_mes_dia.month
        dia = ano_mes_dia.day
        #print(arquivo)
        #print(ano_mes_dia,ano,mes,dia)
        src_dir=os.path.join(caminho,arquivo)
        dst_dir=os.path.join(path_mural,str(ano),str(mes),str(dia),arquivo) 
        print(src_dir)
        print(dst_dir)        
        
        try:

            shutil.copy(src_dir,dst_dir)
        except:
            erro.append(arquivo)
    print(f"A movimentação de arquivos terminou. {len(erro)} erros aconteceram") 
```

```python
movimentarFotosPastas()
```
